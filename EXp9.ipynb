```python
from gensim.models import Word2Vec
```

```python
import requests

url = "https://en.wikipedia.org/wiki/Global_warming"

headers = {
    "user-Agent": "Mozilla/5.0"
}
response = requests.get(url, headers=headers)

source = response.text

print("page load successfully")
```

```python
source
```

```python
soup = bs.BeautifulSoup(source, 'lxml')
```

```python
text =""
for paragraph in soup.find_all('p'):
  text +=paragraph.text
```

```python
text = re.sub(r'\[[0-9]*\]','',text)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d','',text)
# text = re.sub(r'\s+','',text)  # Commented out: This line was concatenating words
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
sentences
```

```python
model = Word2Vec(sentences,vector_size=100,min_count=1)
words = model.wv.index_to_key
```

```python
print(words)
```

```python
vector = model.wv['around']
```

```python
similar = model.wv.most_similar('around')
```
