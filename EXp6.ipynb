```python
import spacy
from nltk.stem import PorterStemmer
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Apple is looking at buying a UK-based startup for $1 billion in London next year."
```

```python
doc = nlp(text)
```

```python
print("Original Text:")
print(text)
```

```python
print("\n1. Tokenization:")
tokens = [token.text for token in doc]
print(tokens)
```

```python
print("\n2. Lemmatization:")
lemmas = [token.lemma_ for token in doc]
print(lemmas)
```

```python
print("\n3. POS Tagging:")
for token in doc:
    print(f"{token.text:12} --> {token.pos_}")
```

```python
print("\n4. Named Entity Recognition:")
for ent in doc.ents:
    print(f"{ent.text:25} --> {ent.label_}")
```

```python
print("\n5. Stop-word Removal:")
filtered_words = [token.text for token in doc if not token.is_stop and not token.is_punct]
print(filtered_words)
```
