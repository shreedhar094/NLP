```python
import nltk
nltk.download('punkt_tab')
```

```python
para = """Bijapur (officially Vijayapura) is the district headquarters of the Bijapur district of the Karnataka state of India. It is also the headquarters for Bijapur Taluk. Bijapur city is well known for its historical monuments of architectural importance built during the rule of the Adil Shahi dynasty. It is also well known for the popular Karnataka premier league team, the Bijapur Bulls. Bijapur is located 519 km (322 mi) northwest of the state capital Bangalore and about 550 km (340 mi) from Mumbai and 210 km (130 mi) north east of the city of Belgaum."""
```

```python
sentence = nltk.sent_tokenize(para)
```

```python
sentence
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
stemmer = PorterStemmer()
```

```python
for i in range(len(sentence)):
 words = nltk.word_tokenize(sentence[i])
 words = [stemmer.stem(word) for word in words]
 sentence[i] = ''.join(words)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
para1 = """Bijapur (officially Vijayapura) is the district headquarters of the Bijapur district of the Karnataka state of India. It is also the headquarters for Bijapur Taluk. Bijapur city is well known for its historical monuments of architectural importance built during the rule of the Adil Shahi dynasty. It is also well known for the popular Karnataka premier league team, the Bijapur Bulls. Bijapur is located 519 km (322 mi) northwest of the state capital Bangalore and about 550 km (340 mi) from Mumbai and 210 km (130 mi) north east of the city of Belgaum."""
```

```python
sentence1 = nltk.sent_tokenize(para1)
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
for i in range(len(sentence1)):
 words = nltk.word_tokenize(sentence1[i])
 words = [lemmatizer.lemmatize(word) for word in words]
 sentence1[i] = ''.join(words)
```

```python
sentence1
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
tag_word = nltk.pos_tag(words)
```

```python
tag_word
```

```python
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
named_Ent = nltk.ne_chunk(tag_word)
```
